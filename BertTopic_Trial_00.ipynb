# Bert Topic 
#### A topic Modelling analyis of 6000 articles from the gov.uk pertaining to the environment.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

# Load text data from output.txt
with open('Unstructured_Corpora.txt', 'r') as f:
    text_data = [line.rstrip() for line in f.readlines()]
```

```python
import bertopic
import pandas as pd

# Create a DataFrame with the documents
df = pd.DataFrame(text_data, columns=['documents'])

# Create a Bertopic model
model = bertopic.BERTopic()

# Fit the model to the documents
topics, _ = model.fit_transform(df['documents'])

# Visualize the topics
model.visualize_topics()

```

```python
# Get the top 5 topics for each document
df['topic'] = topics
df['top_5_topics'] = df['topic'].apply(lambda x: model.get_topic(x)[:5])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import bertopic
import pandas as pd

np.random.seed(0)

# Load text data from output.txt
with open('output.txt', 'r') as f:
    text_data = [line.rstrip() for line in f.readlines()]

# Create a DataFrame with the documents
df = pd.DataFrame(text_data, columns=['documents'])

# Create a Bertopic model
model = bertopic.BERTopic()

# Fit the model to the documents
topics, _ = model.fit_transform(df['documents'])

# Get the top 5 topics for each document
df['topic'] = topics
df['top_5_topics'] = df['topic'].apply(lambda x: model.get_topic(x)[:5])
```

```python
# Generate a word cloud for each of the first 20 topics
for i in range(20):
    print('topic{0}'.format(i))
    topic_words = [t[0] for t in model.get_topic(i)[:10]] # change 10 to however many words you want in the word cloud
    topic_freq_dict = dict(zip(topic_words, [1]*len(topic_words)))
    wordcloud = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(topic_freq_dict)
    plt.figure(figsize=(8,8))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad=0)
    plt.show()
```

### Plot Topics

### Extract Topic distribution

```python
# Load topic distributions using little_mallet_wrapper and assign it to topic_distributions
topic_distributions = None 
topic_distributions = little_mallet_wrapper.load_topic_distributions(path_to_topic_distributions)

# Assign the topic_distributions to a new column 'topic_distributions' in the DataFrame df
df['topic_distributions'] = topic_distributions

# Apply pd.Series to the 'topic_distributions' column in df and assign it to topic_distributions_df
topic_distributions_df = df['topic_distributions'].apply(pd.Series)

# Set the column names of topic_distributions_df to the first four words of each topic, joined by a space
topic_distributions_df.columns = [" ".join(topic[:4]) for topic in topics]

# Concatenate df and topic_distributions_df along the column axis and assign it to trump_df
topic_distributions_df = pd.concat([df['date'], topic_distributions_df], axis=1)
```

```python
topic_distributions_df
```

### Transform the dates

```python
# Convert the index of topic_distributions_df to a DatetimeIndex
topic_distributions_df.index = pd.to_datetime(topic_distributions_df['date'])

# Convert the 'date' column to year-month periods
topic_distributions_df['year-month'] = topic_distributions_df.index.to_period('M')

# Convert each year-month period to a timestamp and assign it to the 'Date (by month)' column
topic_distributions_df['Date (by month)'] = topic_distributions_df['year-month'].dt.to_timestamp()

# Set the 'Date (by month)' column as the index of the DataFrame topic_distributions_df
topic_distributions_df = topic_distributions_df.set_index('Date (by month)')

# Set the 'Date (by month)' column as the index of the DataFrame df
#topic_distributions_df= topic_distributions_df.set_index('Date (by month)')
```

### Group by dates & Normalize

```python
grouped_df = topic_distributions_df.reset_index().drop(columns = ['date','Date (by month)'])
grouped_df
```

```python
columns =list(topic_distributions_df.columns)
for i in ['date','year-month']:
    columns.remove(i)
columns
```

```python
# Group the entries by the unique values in the 'data' column and calculate the sum of the scores for each group
grouped_df = grouped_df.groupby('year-month').sum()

# Count the number of instances for each unique date in the 'data' column and store the counts in a new 'count' column
grouped_df['count'] = topic_distributions_df['year-month'].value_counts()

# Calculate normalized scores for each column individually
for column in columns:
    grouped_df[column + '_normalized'] = grouped_df[column] / grouped_df['count']
```

### Plot the topics

```python
import matplotlib.pyplot as plt

# Select the topics you want to plot (replace with your desired topics)
topic1 = 'waste environment agency fishing_normalized'
topic2 = 'government climate change new_normalized'
topic3 = 'england natural environment nature'

grouped_df.index = grouped_df.index.astype(str)

#grouped_df = grouped_df.loc['2015-04':]

# Create three separate subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 18))

# Plot topic1 in the first subplot
ax1.plot(grouped_df.index, grouped_df[topic1])
ax1.set_title(topic1)
ax1.set_xlabel('Year-Month')
ax1.set_ylabel('Normalized Score')

# Set x-axis label to display every 6th month
ax1.set_xticks(grouped_df.index[::10])

# Plot topic2 in the second subplot
ax2.plot(grouped_df.index, grouped_df[topic2])
ax2.set_title(topic2)
ax2.set_xlabel('Year-Month')
ax2.set_ylabel('Normalized Score')

# Set x-axis label to display every 6th month
ax2.set_xticks(grouped_df.index[::10])

# Plot topic3 in the third subplot
ax3.plot(grouped_df.index, grouped_df[topic3])
ax3.set_title(topic3)
ax3.set_xlabel('Year-Month')
ax3.set_ylabel('Normalized Score')

# Set x-axis label to display every 6th month
ax3.set_xticks(grouped_df.index[::10])

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

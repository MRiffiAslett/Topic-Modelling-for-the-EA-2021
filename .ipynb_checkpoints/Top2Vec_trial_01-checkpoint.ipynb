Update on our text corpus gathering: we've successfully scraped 100 articles from the environmental category of the UK government website. However, when we ran top2vec to extract topics from the corpus, we encountered a challenge. Despite having a diverse set of articles, we found that only two topics emerged, which seemed to have an overlap. This could be due to several reasons, such as the limited scope of the articles or the similarity in the language used to discuss environmental issues. We'll continue to explore ways to improve the topic modeling and refine our corpus to capture a broader range of perspectives on this important topic.

```python
from top2vec import Top2Vec
import matplotlib.pyplot as plt

# Load text data from output.txt
with open('output.txt', 'r') as f:
    text_data = [line.rstrip() for line in f.readlines()]

# Instantiate Top2Vec model
model = Top2Vec(text_data, embedding_model='universal-sentence-encoder')
```

```python
model.get_num_topics()
```

```python
topic_words, word_scores, topic_nums = model.get_topics(model.get_num_topics())
for topic in topic_nums:
    model.generate_topic_wordcloud(topic)
```

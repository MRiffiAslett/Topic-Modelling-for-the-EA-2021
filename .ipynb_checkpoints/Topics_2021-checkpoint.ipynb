## Load libraries

```python
import pandas as pd
import json
import re
```

### Load data

```python
with open('updated_articles.json', 'r') as f:
    data = f.read()
    
df = pd.DataFrame(json.loads(data))

# their is one instance where the date didn't show so we drop it
df.drop(df.loc[df['date'] == "Date not found"].index, inplace = True)
```

```python
df.head(3)
```

### Check the format of each one of them

```python
# the string is in this format "24 March 2023" so we check that each of them follow the same pattern

# Define the pattern for three words separated by a space
pattern = r'^\w+\s+\w+\s+\w+$'

# Create a boolean mask to identify dates that don't match the desired format
mask = ~df['date'].str.match(pattern, na=False, flags=re.IGNORECASE)

# Count the number of instances where the dates don't match the desired format
count = mask.sum()

print(f"Number of instances with dates not in the desired format: {count}")
```

### Transformation

```python
df['date'] = pd.to_datetime(df['date'], format='%d %B %Y', errors='coerce')
```

### Exploration

```python
print(f"the earliest date is {df['date'].min()} and the latest date is {df['date'].max()}")
```

```python
year_counts = df['date'].dt.year.value_counts()

# Print the count of instances for each year
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already obtained the year_counts as shown earlier
# Plotting the counts for each year
year_counts.plot(kind='bar', figsize=(10, 6))

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count of Instances for Each Year')

# Displaying the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already obtained the year_counts as shown earlier
# Sort the year_counts series by index (year)
year_counts_sorted = year_counts.sort_index()

# Plotting the counts for each year
year_counts_sorted.plot(kind='bar', figsize=(10, 6))

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count of Instances for Each Year')

# Displaying the plot
plt.show()
```

```python
import pandas as pd
from top2vec import Top2Vec
import matplotlib.pyplot as plt

# Step 5: Initialize and fit Top2Vec model
model = Top2Vec(documents= list(df['content']), speed='fast-learn', workers=4)
# You can adjust the 'speed' parameter based on your preference and available resources
```

```python
# Step 6: Get the top topics for each time period
df['year'] = pd.to_datetime(df['date']).dt.year
unique_years = df['year'].unique()

top_topics_evolution = []
for year in unique_years:
    documents = df[df['year'] == year]['content']
    topics = model.get_topics(num_topics=3)
    topic_probabilities = topics[1].T  # Transpose the array to match shape
    top_topics_evolution.append(topic_probabilities)

# Step 7: Visualize the topic evolution
topic_labels = ['Topic 1', 'Topic 2', 'Topic 3']
```

```python
plt.figure(figsize=(8, 6))

for j in range(len(topic_labels)):
    topic_probabilities = [top_topics_evolution[i][j][0] for i in range(len(unique_years))]
    plt.plot(unique_years, topic_probabilities, label=topic_labels[j])

plt.xlabel('Year')
plt.ylabel('Topic Probability')
plt.title('Evolution of Top Topics')
plt.legend(title='Topics')
plt.show()
```

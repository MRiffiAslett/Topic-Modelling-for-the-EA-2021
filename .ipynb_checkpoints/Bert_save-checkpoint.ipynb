```python
import pandas as pd
import json

with open('updated_articles.json', 'r') as f:
    data = f.read()
    
df = pd.DataFrame(json.loads(data))

# their is one instance where the date didn't show so we drop it
df.drop(df.loc[df['date'] == "Date not found"].index, inplace = True)

df['content'] = df['content'].astype(str)

# Convert the index of topic_distributions_df to a DatetimeIndex
df.index = pd.to_datetime(df['date'])

# Convert the 'date' column to year-month periods
df['year-month'] = df.index.to_period('M')

# Convert each year-month period to a timestamp and assign it to the 'Date (by month)' column
df['Date (by month)'] = df['year-month'].dt.to_timestamp()

# Set the 'Date (by month)' column as the index of the DataFrame topic_distributions_df
df = df.set_index('Date (by month)')
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the set of stop words the first time
nltk.download('punkt')
nltk.download('stopwords')

# Get English stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words
def remove_stopwords(text):
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word.casefold() not in stop_words]
    return ' '.join(filtered_text)

# Apply the function to the 'content' column
df['content'] = df['content'].apply(remove_stopwords)
```

```python
from bertopic import BERTopic

topic_model = BERTopic(verbose=True)
topics, probs = topic_model.fit_transform(df['content'])
```

```python
topic_model.visualize_topics()
```

```python
timestamps = [str(period) for period in df['year-month']]

content = df.content.to_list()


topics_over_time = topic_model.topics_over_time(content, timestamps, nr_bins=20)
```

```python
topic_model.visualize_topics_over_time(topics_over_time, topics=[1,2,3])
```

```python
topic_model.visualize_topics_over_time(topics_over_time, topics=[4,5,6])
```

# CoherenceScore

```python
import pandas as pd
import re

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_20newsgroups

from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary

Topics = topic_model.get_topic_info()['Representation'].to_list()[1:11][:10]

tokenizer = lambda s: re.findall( '\w+', s.lower() )
texts = [ tokenizer(t) for t in  df['content'] ]

word2id = Dictionary( texts )

# Coherence model
cm = CoherenceModel(topics=Topics, 
                    texts=texts,
                    coherence='c_v',  
                    dictionary=word2id)

coherence_per_topic = cm.get_coherence_per_topic()
```

```python
topics_str = [ '\n '.join(t)[:50] for t in Topics ]
data_topic_score = pd.DataFrame( data=zip(topics_str, coherence_per_topic), columns=['Topic', 'Coherence'] )
data_topic_score = data_topic_score.set_index('Topic')

fig, ax = plt.subplots( figsize=(10,25) )
ax.set_title("Topics coherence\n $C_v$")
sns.heatmap(data=data_topic_score, annot=True, square=True,
            cmap='Reds', fmt='.2f',
            linecolor='black', ax=ax )
plt.yticks( rotation=0 )
ax.set_xlabel('')
ax.set_ylabel('')
fig.show()
```
